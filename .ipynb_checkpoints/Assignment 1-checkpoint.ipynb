Strange atractor

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
x = []
y = []
z = []
for line in open('Strange1.txt', 'r'):
    t, a, b, c  = line.split()
    x.append(float(a))
    y.append(float(b))
    z.append(float(c))
ax.scatter3D(x, y, z, c=z, cmap='Greens');
plt.show()
```

```python
x = []
y = []
for line in open('Lyapunov.txt', 'r'):
    a, b = line.split()
    x.append(float(a))
    y.append(float(b))
np.asarray(x);
np.asarray(y);
```

```python
plt.plot(x,y)
```

```python
from scipy import stats
res = stats.linregress(x[300:], y[300:])
plt.plot(x[300:], y[300:], 'o', label='original data')
plt.plot(x[300:], res.intercept + res.slope*np.double(x[300:]), 'r', label='fitted line')
plt.legend()
plt.show()
print(res.slope)
```

```python
x = []
y = []
for line in open('Strange_geometry.txt', 'r'):
    a, b = line.split()
    x.append(float(a))
    y.append(float(b))
np.asarray(x);
np.asarray(y);
```

```python
plt.plot(x,y)
```

```python
from scipy import stats
res = stats.linregress(x, y)
plt.plot(x, y, 'o', label='original data')
plt.plot(x, res.intercept + res.slope*np.double(x), 'r', label='fitted line')
plt.legend()
plt.show()
print(res.slope)
```
